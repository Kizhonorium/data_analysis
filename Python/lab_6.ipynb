```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

<p>1. Зададим входные данные</p>

```python
x1 = np.array([65.1, 66.5, 66.3, 64.8, 68.4, 67.3,
              70.0, 74.4, 75.5, 80.5, 65.8, 68.5,
              67.5, 70.1, 74.3, 70.4, 74.1, 75.9, 80.5])
x2 = np.array([37.2, 25.9, 36.2, 37.7, 39.0, 36.9,
              39.5, 40.0, 40.5, 39.2, 37.6, 39.1,
              36.8, 39.6, 40.1, 39.1, 40.7, 41.5, 39.2])
y = np.array([21.0, 23.2, 25.0, 24.2, 23.4, 25.5, 
             26.6, 20.1, 28.0, 29.3, 24.3, 23.3,
             25.4, 26.8, 20.2, 24.6, 20.8, 28.4, 29.3])

data = pd.DataFrame({
    'Y': y,
    'X1': x1,
    'X2': x2
})
```

```python
data
```

```python
sns.pairplot(data, kind='reg')
```

```python
data.corr()
```

```python
model = LinearRegression()
model.fit(data[['X1','X2']], data['Y'])
a = model.intercept_
b1, b2 = model.coef_
print('Itercept:', a)
print('Coeficient:', b1, b2)
```
