<p>1. Устройство  состоит  из  <i>5</i> (<i>n=5</i> – номер  студента по списку) независимо работающих  элементов.  Вероятность отказа каждого элемента в одном опыте равна  0,1. Составить закон распределения числа отказавших элементов в одном опыте.</p>
<h4 align='center'><b>Решение</b></h4>
<p>Дискретная случайная величина имеет слудеющие возможные значения:0, 1, 2, 3, 4, 5 (количество отказавших деталей). Така как, отказы элементов независимы один от другого, вероятности отказа каждого элемента равны между собой, следовательно можем применить формулу Бернулли.</p>


```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pymc3 as pm
from scipy import stats
from math import factorial
import arviz as az
plt.rcParams['figure.figsize'] = (10, 6)
SEED = 12
```

```python
def num_of_successes(n:int, k:int) -> int:
    return factorial(n) / (factorial(k) * factorial(n - k))

def probability_of_success(p:float, n:int, k:int) -> float:
    C_kn = num_of_successes(n, k)
    return C_kn * (p ** k) * (1 - p) ** (n - k)
```

```python
n = 5
p = 0.1 
data = list(range(n + 1))
data_of_success = [
    probability_of_success(p=p, n=n, k=event) \
     for event in data
]
```

```python
f, ax = plt.subplots()
ax.vlines(
    range(n + 1), 0, data_of_success, color='b', lw=5
)
ax.set_ylim(0, 1)
ax.plot(
    0, 0, label="n = {:3.2f}\np ={:3.2f}".format(n, p),
    alpha=0
)
ax.legend(fontsize=10)
```

<p>2. В партии 10% нестандартных деталей. Наудачу отобраны <i>n</i> (<i>n</i> – номер студента  по  списку) деталей. Написать биномиальный закон распределения дискретной случайной величины X - числа нестандартных деталей среди отобранных.</p>
