<h4 align='center'><b>Нелинейные модели  парной регрессии</b></h4>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.stats.stattools import durbin_watson
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
x = np.array([9.35, 16.72, 21.87, 27.74, 32.95, 38.97,
             45.87, 52.32, 59.32, 65.57, 75.75])
y = np.array([7.67, 11.83, 13.76, 15.48, 16.99, 17.74,
             18.41, 19.18, 19.88, 20.5, 21.34])
data = pd.DataFrame({
    'X': x,
    'Y': y,
})
data
```

```python
# сделаем замену  z = sqrt(x)
z = np.sqrt(x).reshape(-1, 1)
z
```

```python
y = data['Y']
model = LinearRegression().fit(z, y)
```

```python
b0 = model.intercept_
b1 = model.coef_[0]
print('Intercept:', b0)
print('z:', b1)
```

```python
plt.scatter(z.reshape(1, -1), y, color='blue')
plt.xlabel('переменная х')
plt.ylabel('переменная y')
```

<p>Уравнение регрессии имеет вид:</p>$$ y=2.2832z+ 2.5896$$
<p>Перейдем к начальной замене и получим уравнение нелинейной регрессионной модели:</p> $$ y=2.2832 \sqrt{x}+ 2.5896 $$

<p>Создадим новый вектор со значениями от 1 до 100 для построения графика экспериментальных значений и кривой регрессии:</p>

```python
def mk_model(b0, b1):
    return lambda x: b1 * np.sqrt(x) + b0 

x1 = np.array(range(1, 100))
plt.scatter(data['X'], y, color='blue')
plt.plot(x1, mk_model(b0, b1)(x1), color='red')
plt.xlabel('переменная х')
plt.ylabel('переменная y')
plt.legend(['кривая регрессии', 'Данные выборки'])
```

```python
# нормальность
pp_x = sm.ProbPlot(data['X'], fit=True)
pp_y = sm.ProbPlot(data['Y'], fit=True)
fig = pp_x.qqplot(line='45', other=pp_y)
plt.show()
```
